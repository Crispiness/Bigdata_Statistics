## Random Forest 분석을 위해, X변수와, Y변수로 나누어줌

```python
df_raw_y = df_raw['CONCRETE_COMPRESSIVE_STRENGTH']
df_raw_x = df_raw.drop("CONCRETE_COMPRESSIVE_STRENGTH", axis = 1, inplace = False)
```

## Train/Test/Validation 테스트/최종 테스트를 위해 데이터를 0.3/0.4/0.3 비중으로 나누어줌

```python
df_train_x, df_test_x, df_train_y, df_test_y = train_test_split(df_raw_x, df_raw_y, test_size = 0.3, random_state=42)
df_train_x, df_val_x, df_train_y, df_val_y = train_test_split(df_train_x, df_train_y, test_size = (0.3/0.7), random_state=42)
print("train data X size : {}".format(df_train_x.shape))
print("train data Y size : {}".format(df_train_y.shape))
print("test data X size : {}".format(df_test_x.shape))
print("test data Y size : {}".format(df_test_y.shape))
print("validation data X size : {}".format(df_val_x.shape))
print("validation data Y size : {}".format(df_val_y.shape))
```

## 전처리 끝난 Train Data를 이용해 최초 모델링 실시 (랜덤 state는 1234로 지정)

```python
rf_uncustomized = RandomForestRegressor(random_state=42)
rf_uncustomized.fit(df_train_x, df_train_y)

print("Score on training set : {:.3f}".format(rf_uncustomized.score(df_train_x, df_train_y)))
print("Score on validation set : {:.3f}".format(rf_uncustomized.score(df_val_x, df_val_y)))
```

## n_estimators 조정해가며 모델링 최적화 실시 (10~200 시도)

```python
train_score = []; test_score = []

para_n_tree = [n_tree * 10 for n_tree in range (1,21)]

for v_n_estimators in para_n_tree :
    rf = RandomForestRegressor(random_state = 42
                               ,n_estimators=v_n_estimators)
    rf.fit(df_train_x, df_train_y)
    train_score.append(rf.score(df_train_x, df_train_y))
    test_score.append(rf.score(df_val_x, df_val_y))
    
df_score_n = pd.DataFrame()
df_score_n["n_estimators"] = para_n_tree
df_score_n["TrainScore"] = train_score
df_score_n["TestScore"] = test_score


df_score_n["D"] = df_score_n["TrainScore"]-df_score_n["TestScore"]

df_score_n.round(3)
```

## 아래 그래프로 판단했을때, 모델이 Test 샘플을 설명하는 설명력이 가장 높은 지점인 n_estimator 수를 160으로 설정

```python
plt.plot(para_n_tree, train_score, linestyle = '-', label = 'Train Score')
plt.plot(para_n_tree, test_score, linestyle = '--', label = "Test Score")
plt.ylabel("score"); plt.xlabel("n_estimators")
plt.legend()
```

## Leaf 조절해가며 Decision Tree 모델링 실시 (Leaf 1~30 시도)

```python
train_score = []; test_score = []

para_leaf = [n_leaf * 1 for n_leaf in range (1,31)]

for v_min_samples_leaf in para_leaf :
    rf = RandomForestRegressor(random_state = 42, n_estimators = 160,
                               min_samples_leaf=v_min_samples_leaf)
    rf.fit(df_train_x, df_train_y)
    train_score.append(rf.score(df_train_x, df_train_y))
    test_score.append(rf.score(df_val_x, df_val_y))
    
df_score_n = pd.DataFrame()
df_score_n["MinSamplesLeaf"] = para_leaf
df_score_n["TrainScore"] = train_score
df_score_n["TestScore"] = test_score


df_score_n["D"] = df_score_n["TrainScore"]-df_score_n["TestScore"]

df_score_n.round(3)
```

## 아래 그래프로 판단했을때, 모델이 Test 샘플을 설명하는 설명력이 가장 높은 지점인 최소 Leaf 샘플 수를 1로 설정

```python
plt.plot(para_leaf, train_score, linestyle = '-', label = 'Train Score')
plt.plot(para_leaf, test_score, linestyle = '--', label = "Test Score")
plt.legend()
```

## Leaf 설정이 끝난 후, 노드의 크기를 설정할 차례, 2부터 60 까지 변경해가며 모델링 실시 

```python
train_score = []; test_score = []

para_split = [n_split * 2 for n_split in range(1, 31)]

for v_min_samples_split in para_split :
    rf = RandomForestRegressor(random_state=42, n_estimators = 160, min_samples_leaf=1,
                                min_samples_split = v_min_samples_split)
    rf.fit (df_train_x, df_train_y)
    train_score.append(rf.score(df_train_x, df_train_y))
    test_score.append(rf.score(df_val_x, df_val_y))
    
df_score_split = pd.DataFrame()
df_score_split["MinSampleSplit"] = para_split
df_score_split["TrainScore"] = train_score
df_score_split["TestScore"] = test_score


df_score_split["D"] = df_score_split["TrainScore"]-df_score_split["TestScore"]

df_score_split.round(3)
```

## 아래 그래프로 미뤄보아, 모델이 Test 샘플을 설명하는 설명력이 높으면서 차이가 줄어드는 지점인 최소 노드의 샘플 수를 4로 설정

```python
plt.plot(para_split, train_score, linestyle = '-', label = 'Train Score')
plt.plot(para_split, test_score, linestyle = '--', label = "Test Score")
plt.legend()
```

## Leaf, 노드 샘플 크기 설정이 끝난 후, 깊이를 설정할 차례, 1부터 20 까지 변경해가며 모델링 실시

```python
train_score = []; test_score = []

para_depth = [depth for depth in range(1, 21)]

for v_max_depth in para_depth :
    rf = RandomForestRegressor(random_state=42, n_estimators = 160, min_samples_leaf=1, min_samples_split =2,
                                max_depth = v_max_depth)
    rf.fit (df_train_x, df_train_y)
    train_score.append(rf.score(df_train_x, df_train_y))
    test_score.append(rf.score(df_val_x, df_val_y))
    
df_score_depth = pd.DataFrame()
df_score_depth["Depth"] = para_depth
df_score_depth["TrainScore"] = train_score
df_score_depth["TestScore"] = test_score


df_score_depth["D"] = df_score_depth["TrainScore"]-df_score_depth["TestScore"]

df_score_depth.round(3)
```

## 아래 그래프 확인 시, Depth 14에서 모델이 Test 샘플을 설명하는 설명력이 가장 높게 나타남

```python
plt.plot(para_depth, train_score, linestyle = '-', label = 'Train Score')
plt.plot(para_depth, test_score, linestyle = '--', label = "Test Score")
plt.legend()
```

## 최종 모델을 위에서 확인한 것과 같이, 아래 식으로 설정함 (n_estimator 160 / leaf 1 / split 2 / depth 14)

## Train Data 예측 확률은 97.6%, Validation Data 예측 확률은 90.1%이며, 최종적으로 Test Data를 이용해 최종적으로 모델링을 검증해 보니, 설명력은 85.1%로 나타남

## 콘크리트 강도를 예측하는 모델의 정확성은 약 85.1% 수준이며, Decision Tree분석 대비 모델링의 설명력이 좋게 나타난 것으로 판단되어, 더 적합한 모델이라고 판단된다. 하지만 여전히 15%수준의 설명 예측 불가능한 부분이 존재하여, 모델링을 조금 더 강화할 필요가 있으며, 콘크리트의 강도의 예측 정확성이 치명적인 영향을 미치지 않는다면 해당 모델링을 사용해도 무방할 것으로 판단된다.

```python
rf_final = RandomForestRegressor(random_state = 42, n_estimators = 160, min_samples_leaf=1, min_samples_split =2, max_depth=14)
rf_final.fit(df_train_x, df_train_y)

print("Score on training set: {:.3f}".format(rf_final.score(df_train_x,df_train_y)))
print("score on validation seg: {:.3f}".format(rf_final.score(df_val_x,df_val_y)))
print("score on test seg: {:.3f}".format(rf_final.score(df_test_x,df_test_y)))
```

## 최종 Random Forest 0번의 그래프 모양은 아래와 같다

```python
v_feature_name = df_train_x.columns

export_graphviz(rf_final.estimators_[0], out_file="rfr_final_0.dot", feature_names=v_feature_name, impurity = True, filled = True)
with open("rfr_final_0.dot") as f :
    dot_graph = f.read()

display(graphviz.Source(dot_graph))
```

## 완료된 모델을 이용하여, 테스트 데이터의 예측값을 아래와 같이 계산할 수 있다.

```python
y_predict=rf_final.predict(df_test_x)
y_pre = pd.DataFrame()
y_pre['Predict'] = y_predict

y_pre
```

## 변수별 중요도는 아래와 같이 나타났다

```python
df_importance = pd.DataFrame()
df_importance["Feature"] = v_feature_name
df_importance['Importance'] = rf_final.feature_importances_

df_importance.sort_values("Importance", ascending = False, inplace = True)
df_importance.round(3)
```

```python
df_importance.sort_values("Importance", ascending=True, inplace=True)
coordinates = range(len(df_importance))
plt.barh(y=coordinates, width = df_importance["Importance"])
plt.yticks(coordinates, df_importance['Feature'])
plt.xlabel("변수 중요도")
plt.ylabel("변수")
```

## 모델의 mse와 msa는 아래와 같이 계산할 수 있다

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print("MAE: {:.3f}".format(mean_absolute_error(df_test_y, rf_final.predict(df_test_x))))
print("MSE: {:.3f}".format(mean_squared_error(df_test_y, rf_final.predict(df_test_x))))
```

### MAE는 모델과 각 수치간 평균 절대 오차(|Xi-X바|), MSE는 평균 제곱 오차(|Xi-X바|**2)를 뜻하며, 이 두 수치가 0에 가까울 수록, 모델의 오차가 적다고 볼 수 있다.
