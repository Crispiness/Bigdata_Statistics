```python
df_raw = pd.read_csv("/home/pirl/Statistics/Day3_0715/통신고객이탈.csv", engine = "python", encoding="CP949")
df_raw.head()
```

```python
df_raw.isnull().sum(axis=0)
```

```python
obj_cols = df_raw.select_dtypes("object").columns
for col in obj_cols : 
    _n = len(df_raw[col].value_counts())
    print(col, ":", str(_n))
```

```python
df_raw["CHURN"] = np.where(df_raw['CHURN'] == "Churned",1,0)
df_raw.head()
```

```python
df_raw["CHURN"].sum()
```

```python
df_raw_dummy = pd.get_dummies(df_raw)
df_raw_dummy.head()
```

```python
df_raw_x = df_raw_dummy.drop("CHURN", axis = 1, inplace =False)
df_raw_y = df_raw_dummy["CHURN"]
```

```python
df_train_x, df_test_x, df_train_y, df_test_y = train_test_split(df_raw_x, df_raw_y,
                                                                test_size = 0.3, random_state=1234)
print("train data X size : {}".format(df_train_x.shape))
print("train data Y size : {}".format(df_train_y.shape))
print("test data X size : {}".format(df_test_x.shape))
print("test data Y size : {}".format(df_test_y.shape))
```

```python
para_depth = [depth for depth in range (1,51)]
para_split = [n_split * 10 for n_split in range (1,21)]
para_leaf = [n_leaf * 2 for n_leaf in range (1,21)]
```

```python
estimator = DecisionTreeClassifier()

param_grid = {"max_depth":para_depth, "min_samples_split":para_split,
             "min_samples_leaf":para_leaf}

grid_dt = GridSearchCV(estimator, param_grid, scoring="accuracy", n_jobs = -1)
grid_dt.fit(df_train_x, df_train_y)

print("best estimator model: \n{}".format(grid_dt.best_estimator_))
print("\nbest parameter: \n{}".format(grid_dt.best_params_))
print("\nbest score:\n{}".format(grid_dt.best_score_.round(3)))
```

```python
tree_final = DecisionTreeClassifier(random_state = 1234, max_depth=5, min_samples_split=10, min_samples_leaf=2)
tree_final.fit(df_train_x, df_train_y)
```

```python
v_feature_name = df_train_x.columns

export_graphviz(tree_final, out_file = "tree_final.dot", class_names=["Good", "Bad"]
               ,feature_names=v_feature_name,impurity=True, filled=True)

with open("tree_final.dot") as f :
    dot_graph = f.read()
display(graphviz.Source(dot_graph))
```

```python
y_pred = tree_final.predict(df_test_x)

print("Accuracy: {0:.3f}\n".format(tree_final.score(df_test_x, df_test_y)))
print("Confusion Matrix: \n{}".format(confusion_matrix(df_test_y, y_pred)))
```
