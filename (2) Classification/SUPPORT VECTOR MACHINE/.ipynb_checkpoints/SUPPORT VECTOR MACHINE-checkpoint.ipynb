```python
df_raw = pd.read_csv("/home/pirl/Statistics/Day3_0715/HMEQ.csv", engine = "python", encoding="CP949")
df_raw.head()
```

```python
df_raw.isnull().sum(axis=0)
```

```python
obj_cols = df_raw.select_dtypes("object").columns
for col in obj_cols : 
    _n = len(df_raw[col].value_counts())
    print(col, ":", str(_n))
```

```python
df_raw["JOB"].fillna("Other", inplace=True)
df_raw.fillna(df_raw.mean(), inplace=True)
```

```python
df_raw['JOB'].value_counts()
```

```python
df_raw_dummy = pd.get_dummies(df_raw)
df_raw_dummy.head()
```

```python
df_raw_x = df_raw_dummy.drop("BAD", axis = 1, inplace =False)
df_raw_y = df_raw_dummy["BAD"]
```

```python
df_train_x, df_test_x, df_train_y, df_test_y = train_test_split(df_raw_x, df_raw_y,
                                                                test_size = 0.3, random_state=1234)
print("train data X size : {}".format(df_train_x.shape))
print("train data Y size : {}".format(df_train_y.shape))
print("test data X size : {}".format(df_test_x.shape))
print("test data Y size : {}".format(df_test_y.shape))
```

```python
svm_uncustomized = SVC()
svm_uncustomized.fit(df_train_x, df_train_y)

print("Accuracy on training set: {:.3f}\n".format(svm_uncustomized.score(df_train_x,df_train_y)))
print("Accuracy on test set: {:.3f}\n".format(svm_uncustomized.score(df_test_x,df_test_y)))
```

```python
train_accuracy = []; test_accuracy = []

para_c = [10 ** c for c in range (-2,2)]

for v_C in para_c :
    svm = SVC(C=v_C, random_state=1234)
    svm.fit(df_train_x, df_train_y)
    train_accuracy.append(svm.score(df_train_x, df_train_y))
    test_accuracy.append(svm.score(df_test_x, df_test_y))
    
df_accuracy_c = pd.DataFrame()
df_accuracy_c["C"] = para_c
df_accuracy_c["TrainAccuracy"] = train_accuracy
df_accuracy_c["TestAccuracy"] = test_accuracy


df_accuracy_c["D"] = df_accuracy_c["TrainAccuracy"]-df_accuracy_c["TestAccuracy"]

df_accuracy_c.round(3)
```

```python
plt.plot(para_c, train_accuracy, linestyle = '-', label = 'Train Score')
plt.plot(para_c, test_accuracy, linestyle = '--', label = "Test Score")
plt.legend()
```

```python
train_accuracy = []; test_accuracy = []

para_gamma = [10 ** gamma for gamma in range (-2,2)]

for v_gamma in para_gamma :
    svm = SVC(gamma=v_gamma, C=0.1, random_state=1234)
    svm.fit(df_train_x, df_train_y)
    train_accuracy.append(svm.score(df_train_x, df_train_y))
    test_accuracy.append(svm.score(df_test_x, df_test_y))
    
df_accuracy_gamma = pd.DataFrame()
df_accuracy_gamma["gamma"] = para_gamma
df_accuracy_gamma["TrainAccuracy"] = train_accuracy
df_accuracy_gamma["TestAccuracy"] = test_accuracy


df_accuracy_gamma["D"] = df_accuracy_gamma["TrainAccuracy"]-df_accuracy_gamma["TestAccuracy"]

df_accuracy_gamma.round(3)
```

```python
plt.plot(para_gamma, train_accuracy, linestyle = '-', label = 'Train Score')
plt.plot(para_gamma, test_accuracy, linestyle = '--', label = "Test Score")
plt.legend()
```

```python
train_score = []; test_score = []

para_C = [10 ** c for c in range (-2,2)]

for v_C in para_C :
    svm = SVC(C= v_C)
    svm.fit(df_scaled_train_x,df_train_y)
    train_score.append(svm.score(df_scaled_train_x, df_train_y))
    test_score.append(svm.score(df_scaled_test_x, df_test_y))
    
df_score_C = pd.DataFrame()
df_score_C["C"] = para_C
df_score_C["TrainScore"] = train_score
df_score_C["TestScore"] = test_score


df_score_C["D"] = df_score_C["TrainScore"]-df_score_C["TestScore"]

df_score_C.round(3)
```

```python
plt.plot(para_C, train_score, linestyle = '-', label = 'Train Score')
plt.plot(para_C, test_score, linestyle = '--', label = "Test Score")
plt.ylabel("score"); plt.xlabel("C")
plt.legend()
```

```python
train_score = []; test_score = []

para_gamma = [10 ** gamma for gamma in range (-2,3)]

for v_gamma in para_gamma :
    svm = SVC(gamma = v_gamma, C = 1)
    svm.fit(df_scaled_train_x,df_train_y)
    train_score.append(svm.score(df_scaled_train_x, df_train_y))
    test_score.append(svm.score(df_scaled_test_x, df_test_y))
    
df_score_gamma = pd.DataFrame()
df_score_gamma["gamma"] = para_gamma
df_score_gamma["TrainScore"] = train_score
df_score_gamma["TestScore"] = test_score


df_score_gamma["D"] = df_score_gamma["TrainScore"]-df_score_gamma["TestScore"]

df_score_gamma.round(3)
```

```python
plt.plot(para_gamma, train_score, linestyle = '-', label = 'Train Score')
plt.plot(para_gamma, test_score, linestyle = '--', label = "Test Score")
plt.ylabel("score"); plt.xlabel("gamma")
plt.legend()
```

```python
svc_final = SVC(C = 1, gamma = 0.1)
svc_final.fit(df_scaled_train_x, df_train_y)
```

```python
y_pred = svc_final.predict(df_scaled_test_x)

print("Accuracy on training set: {0:.3f}\n".format(svc_final.score(df_scaled_train_x, df_train_y)))
print("Accuracy on test set: {0:.3f}\n".format(svc_final.score(df_scaled_test_x, df_test_y)))
print("Confusion Matrix: \n{}".format(confusion_matrix(df_test_y, y_pred)))
```
