# 로지스틱 회귀모델 및 회귀계수 검토

```python
df_char = df_raw.select_dtypes(include = "object")
df_numeric = df_raw.select_dtypes(exclude = "object")

scaler = StandardScaler()
np_numeric_scaled = scaler.fit_transform(df_numeric)
df_numeric_scaled = pd.DataFrame(np_numeric_scaled, columns=df_numeric.columns)

df_scaled = pd.concat([df_numeric_scaled, df_char], axis=1)
df_scaled.head()
```

```python
df_scaled["CHURN"] = np.where(df_scaled['CHURN'] == "Churned",1,0)
df_scaled.head()
```

```python
df_scaled["CHURN"].sum()
```

```python
'''
df_scaled.drop("BAD_TEST", inplace = True, axis = 1)
```

```python
df_scaled_train, df_scaled_test = train_test_split(df_scaled, test_size = 0.3, random_state=1234)
print("train data size : {}".format(df_scaled_train.shape))
print("test data size : {}".format(df_scaled_test.shape))
```

```python
scaled_log_model = Logit.from_formula("CHURN ~ C(GENDER)+AGE+C(CHARGE_TYPE)+C(HANDSET)+C(USAGE_BAND)\
                                      +SERVICE_DURATION+DROPPED_CALLS+PEAK_CALLS_NO+PEAK_CALLS_TIME+WEEKEND_CALLS_NO\
                                      +WEEKEND_CALLS_TIME+TOTAL_CALLS_NO+TOTAL_CALLS_TIME", df_scaled_train)

scaled_log_result = scaled_log_model.fit()

print(scaled_log_result.summary())
```

```python
df_logistic_coef=pd.DataFrame({"Coef":scaled_log_result.params.values[1:]}, index=scaled_log_model.exog_names[1:])
df_logistic_coef.plot.barh(y='Coef', color = "green")
```

```python
y_pred = scaled_log_result.predict(df_scaled_test)
y_pred_class = (y_pred > 0.5).astype(int)
y_pred_class.head()
```

```python
y_pred.head()
```

```python
print("Accuracy: {0:.3f}\n".format(accuracy_score(df_scaled_test['CHURN'],y_pred_class)))
print("Confusion Matrix: \n{}".format(confusion_matrix(df_scaled_test['CHURN'],y_pred_class)))
```
